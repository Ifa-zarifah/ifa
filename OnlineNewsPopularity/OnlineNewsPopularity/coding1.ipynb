```python
from sklearn import preprocessing
```

```python
popularity_raw = df[df.keys()[-1]]
popularity_raw.describe()
```

```python
label_encoder = preprocessing.LabelEncoder()
popular_label = pd.Series(label_encoder.fit_transform(popularity_raw>=1400))
```

```python
# Get the features from dataset
features_raw = df.drop(['url','timedelta','is_weekend'], axis=1)
display(features_raw.head())
```

```python
features_raw.shape
```

```python
# Visualize the feature of different day of week
columns_day = features_raw.columns.values[29:36]
unpop=df[df['shares']<1400]
pop=df[df['shares']>=1400]
unpop_day = unpop[columns_day].sum().values
pop_day = pop[columns_day].sum().values
```

```python
import matplotlib.pyplot as pl
from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
fig = pl.figure(figsize = (13,5))
pl.title("Count of popular/unpopular news over different day of week")
pl.bar(np.arange(len(columns_day)), pop_day, width = 0.3, align="center",color = 'crimson', label = "popular")
pl.bar(np.arange(len(columns_day)) - 0.3, unpop_day, width = 0.3, align = "center", color = 'darkblue', label = "unpopular")
pl.xticks(np.arange(len(columns_day)), columns_day)
pl.ylabel("Count")
pl.xlabel("Days of week")
    
pl.legend(loc = 'upper right')
pl.tight_layout()
pl.savefig("days.pdf")
pl.show()
```

```python
plt.pie(pop_day,labels=columns_day,startangle=90, autopct='%.1f%%')
plt.title('Popular news over different day of week')
plt.legend(bbox_to_anchor=(1.45,1.025),loc="upper left")
plt.show()
```

```python
plt.pie(unpop_day,labels=columns_day,startangle=90, autopct='%.1f%%')
plt.title('Unpopular news over different day of week')
plt.legend(bbox_to_anchor=(1.45,1.025),loc="upper left")
plt.show()
```

```python
# Normalize the numerical features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = ['n_tokens_title', 'n_tokens_content', 'num_hrefs', 'num_self_hrefs', 'num_imgs','num_videos','average_token_length',
             'num_keywords','self_reference_min_shares','self_reference_max_shares','self_reference_avg_sharess']
features_raw[numerical] = scaler.fit_transform(df[numerical])
display(features_raw.head(n = 1))
```

```python
from sklearn.feature_selection import RFECV
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
ada = AdaBoostClassifier(random_state=0)
rfecv = RFECV (estimator = ada,step=1, cv=5, scoring ='accuracy')
selector_ada= rfecv.fit (features_raw, popular_label)
selector_ada.ranking_

print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()

print (features_raw.columns.values[selector_ada.ranking_==1].shape[0])
print (features_raw.columns.values[selector_ada.ranking_==1])
features_ADA = features_raw[features_raw.columns.values[selector_ada.ranking_==1]]

```

```python
estimator = AdaBoostClassifier(random_state=0, n_estimators=100)
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(features_raw, popular_label)
selector.support_
selector.ranking_

estimator_RF = RandomForestClassifier(random_state=0, n_estimators=100)
selector_RF = RFECV(estimator_RF, step=1, cv=5)
selector_RF = selector_RF.fit(features_raw, popular_label)
selector_RF.support_
selector_RF.ranking_

estimator_GBM = HistGradientBoostingClassifier(random_state=0, n_estimators=100)
selector_GBM = RFECV(estimator_RF, step=1, cv=5)
selector_GBM = selector_GBM.fit(features_raw, popular_label)
selector_GBM.support_
selector_GBM.ranking_
```

```python
# Plot the cv score vs number of features
pl.figure()
pl.xlabel("Number of features selected")
pl.ylabel("Cross validation score (nb of correct classifications)")
pl.plot(range(1, len(selector.grid_scores_) + 1), selector.grid_scores_)
pl.savefig('RFE_ADA.pdf')
pl.show()

print features_raw.columns.values[selector.ranking_==1].shape[0]
print features_raw.columns.values[selector.ranking_==1]
features_ADA = features_raw[features_raw.columns.values[selector.ranking_==1]]

pl.figure()
pl.xlabel("Number of features selected")
pl.ylabel("Cross validation score (nb of correct classifications)")
pl.plot(range(1, len(selector_RF.grid_scores_) + 1), selector_RF.grid_scores_)
pl.savefig('RFE_RF.pdf')
pl.show()

print features_raw.columns.values[selector_RF.ranking_!=1].shape[0]
print features_raw.columns.values[selector_RF.ranking_!=1]
features_RF = features_raw[features_raw.columns.values[selector_RF.ranking_==1]]

pl.figure()
pl.xlabel("Number of features selected")
pl.ylabel("Cross validation score (nb of correct classifications)")
pl.plot(range(1, len(selector_GBM.grid_scores_) + 1), selector_GBM.grid_scores_)
pl.savefig('RFE_GBM.pdf')
pl.show()

print features_raw.columns.values[selector_GBM.ranking_!=1].shape[0]
print features_raw.columns.values[selector_GBM.ranking_!=1]
features_GBM = features_raw[features_raw.columns.values[selector_GBM.ranking_==1]]
```

```python
# Split data into training and testing sets
from sklearn.metrics import accuracy_score, fbeta_score, roc_curve, auc, roc_auc_score
from sklearn.cross_validation import train_test_split

X_train_ADA, X_test_ADA, y_train_ADA, y_test_ADA = train_test_split(features_ADA, popular_label, test_size = 0.1, random_state = 0)

X_train_RF, X_test_RF, y_train_RF, y_test_RF = train_test_split(features_RF, popular_label, test_size = 0.1, random_state = 0)

X_train_GBM, X_test_GBM, y_train_GBM, y_test_GBM = train_test_split(features_GBM, popular_label, test_size = 0.1, random_state = 0)

print "Training set has {} samples.".format(X_train_ADA.shape[0])

print "Testing set has {} samples.".format(X_test_ADA.shape[0])
```
